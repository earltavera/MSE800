```python
#Gross Pay Calculator

def calculate_gross_pay():
    """
    1. INPUT: Receives hours worked and hourly rate.
    2. PROCESS: Calculates the gross pay.
    3. OUTPUT: Displays the result.
    """
    print("--- Gross Pay Calculator ---")

    # --- 1. INPUT ---
    # Get the hours worked from the user. We use float() to handle decimal numbers.
    try:
        hours_worked = float(input("Enter the total number of hours worked: "))
    except ValueError:
        print("Invalid input for the total hours. Please enter a number.")
        return

    # Get the hourly pay rate from the user.
    try:
        hourly_rate = float(input("Enter the hourly pay rate: $"))
    except ValueError:
        print("Invalid input for rate. Please enter a number.")
        return

    # Check for non-negative inputs, we need value above zero
    if hours_worked < 0 or hourly_rate < 0:
        print("Hours and rate must be non-negative.")
        return

    # --- 2. PROCESS ---
    # Multiply the hours worked by the hourly pay rate to get the gross pay.
    gross_pay = hours_worked * hourly_rate

    # --- 3. OUTPUT ---
    # Display the calculated result, formatted to two decimal places.
    print(f"\nProcessing Complete...")
    print(f"Gross Pay: ${gross_pay:.2f}")

# Execute the function to run the program
calculate_gross_pay()
```
